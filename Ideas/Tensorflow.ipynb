```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Dropout, Concatenate
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import yfinance as yahoo_finance  # For market data
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re
import vectorbt as vbt

# Step 1: Data Collection and Preprocessing
# ------------------------------------------

# Function to download and prepare market data
def get_market_data(symbol, start_date, end_date, interval='1d'):
    """
    Download market data for a crypto asset
    """
    # Download data
    data = yahoo_finance.download(symbol, start=start_date, end=end_date, interval=interval)
    
    # Fix MultiIndex columns issue by flattening column names
    if isinstance(data.columns, pd.MultiIndex):
        data.columns = [col[0] for col in data.columns]
    
    # Fix for MultiIndex issue - ensure we have a single-level DatetimeIndex
    if data.index.nlevels > 1:
        data = data.reset_index(level=1, drop=True)
    
    # Calculate additional technical indicators
    data['Returns'] = data['Close'].pct_change()
    data['MA5'] = data['Close'].rolling(window=5).mean()
    data['MA20'] = data['Close'].rolling(window=20).mean()
    
    # Calculate RSI (14-period)
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    data['RSI'] = 100 - (100 / (1 + rs))
    
    # Calculate MACD
    data['EMA12'] = data['Close'].ewm(span=12, adjust=False).mean()
    data['EMA26'] = data['Close'].ewm(span=26, adjust=False).mean()
    data['MACD'] = data['EMA12'] - data['EMA26']
    data['Signal_line'] = data['MACD'].ewm(span=9, adjust=False).mean()
    
    # Calculate volatility (20-day standard deviation of returns)
    data['Volatility'] = data['Returns'].rolling(window=20).std()
    
    # Clean up data
    data = data.dropna()
    
    return data

# Function to process text data and extract sentiment
def preprocess_text_data(text_data):
    """
    Clean and prepare text data for sentiment analysis
    """
    # Initialize NLTK's VADER sentiment analyzer
    nltk.download('vader_lexicon', quiet=True)
    sia = SentimentIntensityAnalyzer()
    
    # Clean the text data
    text_data['clean_text'] = text_data['text'].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', str(x).lower()))
    
    # Get sentiment scores
    text_data['sentiment_compound'] = text_data['clean_text'].apply(lambda x: sia.polarity_scores(x)['compound'])
    text_data['sentiment_pos'] = text_data['clean_text'].apply(lambda x: sia.polarity_scores(x)['pos'])
    text_data['sentiment_neg'] = text_data['clean_text'].apply(lambda x: sia.polarity_scores(x)['neg'])
    text_data['sentiment_neu'] = text_data['clean_text'].apply(lambda x: sia.polarity_scores(x)['neu'])
    
    return text_data

# Function to aggregate daily sentiment
def aggregate_daily_sentiment(text_data):
    """
    Aggregate sentiment scores by date
    """
    # Group by date and calculate average sentiment scores
    daily_sentiment = text_data.groupby('date').agg({
        'sentiment_compound': 'mean',
        'sentiment_pos': 'mean',
        'sentiment_neg': 'mean',
        'sentiment_neu': 'mean',
        'text': 'count'  # Count of daily mentions
    }).rename(columns={'text': 'mention_count'})
    
    return daily_sentiment

# Function to generate simulated sentiment data
def generate_simulated_sentiment(dates, crypto_symbol):
    """
    Generate simulated sentiment data for each date
    """
    np.random.seed(42)
    
    # Create lists to store data
    all_dates = []
    all_texts = []
    
    # For each date, generate a random number of text entries
    for date in dates:
        num_entries = np.random.randint(5, 20)
        date_entries = [date] * num_entries
        all_dates.extend(date_entries)
        
        # Generate text entries
        texts = [f"This is a simulated {'positive' if np.random.random() > 0.5 else 'negative'} text about {crypto_symbol}" 
                for _ in range(num_entries)]
        all_texts.extend(texts)
    
    # Create DataFrame
    simulated_text_data = pd.DataFrame({
        'date': all_dates,
        'text': all_texts
    })
    
    return simulated_text_data

# Step 2: Feature Engineering
# ---------------------------

def prepare_model_data(market_data, sentiment_data, lookback=10):
    """
    Combine market and sentiment data and prepare sequences for the model
    """
    # Ensure both dataframes have single-level DatetimeIndex
    if market_data.index.nlevels > 1:
        market_data = market_data.reset_index(level=1, drop=True)
    
    # Ensure both have DatetimeIndex
    market_data.index = pd.to_datetime(market_data.index)
    sentiment_data.index = pd.to_datetime(sentiment_data.index)
    
    # Make sure index name matches
    sentiment_data.index.name = market_data.index.name
    
    # Debug information
    print(f"Market data columns: {market_data.columns.tolist()}")
    print(f"Market data shape: {market_data.shape}")
    print(f"Sentiment data columns: {sentiment_data.columns.tolist()}")
    print(f"Sentiment data shape: {sentiment_data.shape}")
    
    # Merge market data with sentiment data
    try:
        combined_data = pd.merge(market_data, sentiment_data, left_index=True, right_index=True, how='inner')
        print(f"Merge successful. Combined data shape: {combined_data.shape}")
        print(f"Combined data columns: {combined_data.columns.tolist()}")
    except Exception as e:
        print(f"Error during merge: {e}")
        print("Trying alternative merge method...")
        # Alternative merge method
        combined_data = pd.concat([market_data, sentiment_data], axis=1, join='inner')
        print(f"Alternative merge successful. Combined data shape: {combined_data.shape}")
        print(f"Combined data columns: {combined_data.columns.tolist()}")
    
    # Verify if 'Returns' column exists
    if 'Returns' not in combined_data.columns:
        print("Warning: 'Returns' column not found in combined data. Recalculating...")
        combined_data['Returns'] = combined_data['Close'].pct_change()
    
    # Select features - ensure they exist in the combined data
    market_features = []
    for feature in ['Close', 'Volume', 'Returns', 'MA5', 'MA20', 'RSI', 'MACD', 'Volatility']:
        if feature in combined_data.columns:
            market_features.append(feature)
        else:
            print(f"Warning: Feature '{feature}' not found in combined data!")
    
    sentiment_features = []
    for feature in ['sentiment_compound', 'sentiment_pos', 'sentiment_neg', 'sentiment_neu', 'mention_count']:
        if feature in combined_data.columns:
            sentiment_features.append(feature)
        else:
            print(f"Warning: Feature '{feature}' not found in combined data!")
    
    # Create target: Next day's return (classification: 1 if positive, 0 if negative)
    combined_data['target'] = (combined_data['Returns'].shift(-1) > 0).astype(int)
    
    # Drop rows with NaN values
    combined_data = combined_data.dropna()
    
    # Make sure we have enough data for model training
    if len(combined_data) <= lookback:
        raise ValueError(f"Not enough data after preprocessing. Only {len(combined_data)} rows available but {lookback} required.")
    
    # Scale market features
    market_scaler = MinMaxScaler()
    combined_data[market_features] = market_scaler.fit_transform(combined_data[market_features])
    
    # Scale sentiment features
    sentiment_scaler = MinMaxScaler()
    combined_data[sentiment_features] = sentiment_scaler.fit_transform(combined_data[sentiment_features])
    
    # Create sequences for LSTM
    X_market = []
    X_sentiment = []
    y = []
    
    for i in range(lookback, len(combined_data)):
        X_market.append(combined_data[market_features].values[i-lookback:i])
        X_sentiment.append(combined_data[sentiment_features].values[i-lookback:i])
        y.append(combined_data['target'].values[i])
    
    return np.array(X_market), np.array(X_sentiment), np.array(y), combined_data

# Step 3: Model Architecture
# --------------------------

def create_dual_input_model(market_input_shape, sentiment_input_shape):
    """
    Create a model that processes both market data and sentiment data
    """
    # Market data input branch
    market_input = Input(shape=market_input_shape, name='market_input')
    market_lstm = LSTM(64, return_sequences=True)(market_input)
    market_lstm = Dropout(0.2)(market_lstm)
    market_lstm = LSTM(32)(market_lstm)
    market_output = Dropout(0.2)(market_lstm)
    
    # Sentiment data input branch
    sentiment_input = Input(shape=sentiment_input_shape, name='sentiment_input')
    sentiment_lstm = LSTM(32, return_sequences=True)(sentiment_input)
    sentiment_lstm = Dropout(0.2)(sentiment_lstm)
    sentiment_lstm = LSTM(16)(sentiment_lstm)
    sentiment_output = Dropout(0.2)(sentiment_lstm)
    
    # Combine both branches
    combined = Concatenate()([market_output, sentiment_output])
    
    # Output layers
    x = Dense(32, activation='relu')(combined)
    x = Dropout(0.2)(x)
    main_output = Dense(1, activation='sigmoid', name='main_output')(x)
    
    # Create and compile model
    model = Model(inputs=[market_input, sentiment_input], outputs=main_output)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    return model

# Step 4: Training and Evaluation
# -------------------------------

def train_and_evaluate_model(X_market, X_sentiment, y, test_size=0.2, epochs=50, batch_size=32):
    """
    Train the model and evaluate its performance
    """
    # Split data into training and testing sets
    X_market_train, X_market_test, X_sentiment_train, X_sentiment_test, y_train, y_test = train_test_split(
        X_market, X_sentiment, y, test_size=test_size, shuffle=False
    )
    
    # Create model
    model = create_dual_input_model(X_market.shape[1:], X_sentiment.shape[1:])
    
    # Train model
    history = model.fit(
        [X_market_train, X_sentiment_train], y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=([X_market_test, X_sentiment_test], y_test),
        verbose=1
    )
    
    # Evaluate model
    loss, accuracy = model.evaluate([X_market_test, X_sentiment_test], y_test)
    print(f"Test Loss: {loss:.4f}")
    print(f"Test Accuracy: {accuracy:.4f}")
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.legend()
    plt.title('Loss')
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.legend()
    plt.title('Accuracy')
    
    plt.tight_layout()
    plt.show()
    
    return model, history, X_market_test, X_sentiment_test, y_test

# Step 5: Trading Strategy Implementation with VectorBT
# ----------------------------------------------------
# FIXED to handle different VectorBT versions

def implement_trading_strategy_vectorbt(model, X_market_test, X_sentiment_test, combined_data, test_size=0.2, threshold=0.6):
    """
    Implement and backtest a trading strategy based on model predictions using vectorbt
    """
    # Get predictions
    y_pred_proba = model.predict([X_market_test, X_sentiment_test])
    print(f"Generated {len(y_pred_proba)} predictions with threshold {threshold}")
    
    # Get the dates for the test period
    test_indices = int(len(combined_data) * (1 - test_size))
    test_data = combined_data.iloc[test_indices:].copy()
    print(f"Test data size: {len(test_data)}")
    
    # Debug: Check if we have any test data
    if len(test_data) == 0:
        raise ValueError("Test data is empty! Check your data splitting logic.")
    
    # Create a DataFrame to store predictions with proper date alignment
    test_dates = test_data.index
    
    # Create a properly aligned prediction array
    predictions_df = pd.DataFrame(index=test_dates, columns=['signal_probability'])
    
    # Calculate how many rows to skip at the beginning of test_data
    offset = len(test_data) - len(y_pred_proba)
    
    if offset < 0:
        # Predictions are longer than test data (shouldn't happen normally)
        print("WARNING: More predictions than test data rows!")
        predictions_df['signal_probability'] = y_pred_proba[:len(test_data)].flatten()
    else:
        # Fill the predictions starting from the offset position
        for i in range(len(y_pred_proba)):
            predictions_df.iloc[offset + i, 0] = y_pred_proba[i][0]  # Extract from 2D array
    
    # Forward fill the NaN values with 0 (no signal) 
    predictions_df = predictions_df.fillna(0)
    
    # Merge with test data
    test_data = test_data.join(predictions_df)
    
    # Create binary signals based on threshold
    test_data['signal'] = (test_data['signal_probability'] > threshold).astype(int)
    
    # Debug: Check if we have any signals
    signal_count = test_data['signal'].sum()
    print(f"Generated {signal_count} buy signals out of {len(test_data)} days")
    if signal_count == 0:
        print("WARNING: No buy signals generated! Try lowering the threshold.")
    
    # Ensure we avoid look-ahead bias
    test_data['signal_next_day'] = test_data['signal'].shift(1).fillna(0)
    
    # Extract price data for vectorbt
    close_price = test_data['Close']
    
    # Debug: Show some of the signals
    print("\nSample of signals and prices:")
    signal_sample = test_data[['Close', 'signal_probability', 'signal', 'signal_next_day']].head(10)
    print(signal_sample)
    
    # Create vectorbt Portfolio using the properly lagged signals
    portfolio = vbt.Portfolio.from_signals(
        close_price,
        entries=test_data['signal_next_day'] == 1,  # Properly lagged entry signals
        exits=test_data['signal_next_day'] == 0,    # Properly lagged exit signals
        fees=0.001,  # 0.1% trading fee
        freq='1D'
    )
    
    # Generate performance metrics
    print("\n================ Strategy Performance ================")
    print(f"Total Return: {portfolio.total_return():.2%}")
    print(f"Sharpe Ratio: {portfolio.sharpe_ratio():.2f}")
    print(f"Max Drawdown: {portfolio.max_drawdown():.2%}")
    
    # Handle potential compatibility issues with VectorBT versions
    try:
        win_rate = portfolio.win_rate()
        print(f"Win Rate: {win_rate:.2%}")
    except AttributeError:
        print("Win Rate: Method not available in your VectorBT version")
        
    try:
        profit_factor = portfolio.profit_factor()
        print(f"Profit Factor: {profit_factor:.2f}")
    except AttributeError:
        print("Profit Factor: Method not available in your VectorBT version")
    
    # FIXED: Use completely manual plotting to avoid vectorbt plotting compatibility issues
    # This approach works with all vectorbt versions
    
    # Create our own figures instead of using vectorbt's plotting functions
    fig, axes = plt.subplots(3, 1, figsize=(15, 18))
    
    # Plot 1: Cumulative returns - Manual implementation
    try:
        # Get returns from portfolio
        returns = portfolio.returns()
        cum_returns = (1 + returns).cumprod()
        
        # Buy & Hold for comparison
        buy_hold_rets = close_price.pct_change().fillna(0)
        buy_hold_cum = (1 + buy_hold_rets).cumprod()
        
        # Plot both lines
        axes[0].plot(cum_returns.index, cum_returns.values, 'g-', label='Strategy')
        axes[0].plot(buy_hold_cum.index, buy_hold_cum.values, 'b--', label='Buy & Hold')
        axes[0].set_title('Cumulative Returns: Strategy vs Buy & Hold')
        axes[0].legend()
        axes[0].grid(True)
        
    except Exception as e:
        print(f"Error plotting cumulative returns: {e}")
        
    # Plot 2: Drawdowns - Manual implementation
    try:
        # Calculate drawdowns manually
        returns = portfolio.returns().fillna(0)
        cum_returns = (1 + returns).cumprod()
        running_max = np.maximum.accumulate(cum_returns)
        drawdowns = (cum_returns / running_max) - 1
        
        # Plot drawdowns
        axes[1].fill_between(drawdowns.index, 0, drawdowns.values, color='red', alpha=0.3)
        axes[1].set_title('Strategy Drawdowns')
        axes[1].grid(True)
        axes[1].set_ylabel('Drawdown')
        
    except Exception as e:
        print(f"Error plotting drawdowns: {e}")
        
    # Plot 3: Trade analysis with price data
    # Create an analysis of when trades were executed
    try:
        # Find where signals change
        entry_trades = test_data[test_data['signal_next_day'].diff() == 1]
        exit_trades = test_data[test_data['signal_next_day'].diff() == -1]
        
        # Check if we have any trades to display
        print(f"Entry trades: {len(entry_trades)}")
        print(f"Exit trades: {len(exit_trades)}")
        
        # Plot price line
        axes[2].plot(close_price.index, close_price.values, label='Close Price', alpha=0.7)
        
        # Add entry and exit points
        if len(entry_trades) > 0:
            axes[2].scatter(entry_trades.index, entry_trades['Close'], color='green', 
                            marker='^', s=100, label='Buy Signal')
        if len(exit_trades) > 0:
            axes[2].scatter(exit_trades.index, exit_trades['Close'], color='red', 
                            marker='v', s=100, label='Sell Signal')
        
        # Add sentiment info to the plot if available
        if 'sentiment_compound' in test_data.columns:
            ax2 = axes[2].twinx()
            ax2.plot(test_data.index, test_data['sentiment_compound'], color='purple', 
                     alpha=0.5, label='Sentiment')
            ax2.set_ylabel('Sentiment Score', color='purple')
            ax2.legend(loc='upper right')
        
        axes[2].set_title('Trade Signals with Price and Sentiment')
        axes[2].set_xlabel('Date')
        axes[2].set_ylabel('Price')
        axes[2].legend(loc='upper left')
        
    except Exception as e:
        print(f"Error plotting trade signals: {e}")
    
    plt.tight_layout()
    plt.show()
    
    # FIXED: Handle the FutureWarning for resample('M')
    print("\nSignal distribution by month:")
    try:
        # Use 'ME' instead of 'M' to avoid the FutureWarning
        monthly_signals = test_data.resample('ME')['signal_next_day'].sum()
        print(monthly_signals)
    except Exception as e:
        print(f"Could not calculate monthly signals: {e}")
    
    # Create more detailed trade analysis
    print("\n================ Trade Analysis ================")
    
    # Try to get trade statistics from vectorbt
    try:
        trades = portfolio.trades
        
        if hasattr(trades, 'records') and len(trades.records) > 0:
            print(f"Total trades: {len(trades.records)}")
            
            # Try different attributes depending on vectorbt version
            if hasattr(trades, 'winning'):
                print(f"Winning trades: {len(trades.winning.records)}")
                print(f"Losing trades: {len(trades.losing.records)}")
            
            # Try to print average trade duration
            try:
                avg_duration = trades.duration.mean()
                print(f"Average trade duration: {avg_duration:.1f} days")
            except:
                pass
                
            # Try to print average profit per trade
            try:
                avg_profit = trades.pnl.mean()
                print(f"Average profit per trade: {avg_profit:.2%}")
            except:
                pass
                
            # Try to print win rate
            try:
                win_rate = len(trades.winning.records) / len(trades.records)
                print(f"Win rate: {win_rate:.2%}")
            except:
                pass
        else:
            print("No trades were executed.")
            
    except Exception as e:
        print(f"Could not analyze trades: {e}")
    
    return portfolio, test_data


    
  
# Step 6: Main Execution
# ---------------------

def main():
    # Set parameters
    crypto_symbol = 'BTC-USD'  # Bitcoin
    start_date = '2020-01-01'
    end_date = '2023-12-31'
    lookback_period = 10  # Number of days to look back for predictions
    test_size = 0.2  # Use 20% of data for testing
    
    # Load market data
    print("Loading market data...")
    market_data = get_market_data(crypto_symbol, start_date, end_date)
    print(f"Market data loaded with shape: {market_data.shape}")
    
    # Fix MultiIndex columns issue if it still exists
    if isinstance(market_data.columns, pd.MultiIndex):
        market_data.columns = [col[0] for col in market_data.columns]
        print("Fixed MultiIndex in market data columns")
    
    # For this example, we'll simulate sentiment data
    print("Simulating sentiment data...")
    
    # Generate simulated sentiment data
    simulated_text_data = generate_simulated_sentiment(market_data.index, crypto_symbol)
    
    # Process text data
    processed_text_data = preprocess_text_data(simulated_text_data)
    daily_sentiment = aggregate_daily_sentiment(processed_text_data)
    print(f"Sentiment data processed with shape: {daily_sentiment.shape}")
    
    # Ensure proper index alignment
    market_data.index = pd.to_datetime(market_data.index)
    daily_sentiment.index = pd.to_datetime(daily_sentiment.index)
    
    # If market_data has MultiIndex, flatten it
    if market_data.index.nlevels > 1:
        print("Flattening market data MultiIndex...")
        market_data = market_data.reset_index(level=1, drop=True)
    
    # Ensure sentiment data index name matches market data
    daily_sentiment.index.name = market_data.index.name
    
    # Ensure the two dataframes have compatible indices
    common_dates = market_data.index.intersection(daily_sentiment.index)
    if len(common_dates) < len(market_data.index) or len(common_dates) < len(daily_sentiment.index):
        print(f"Filtering to {len(common_dates)} common dates...")
        market_data = market_data.loc[common_dates]
        daily_sentiment = daily_sentiment.loc[common_dates]
    
    # Prepare model data
    print("Preparing model data...")
    try:
        X_market, X_sentiment, y, combined_data = prepare_model_data(market_data, daily_sentiment, lookback=lookback_period)
        print(f"Model data prepared. X_market shape: {X_market.shape}, X_sentiment shape: {X_sentiment.shape}")
    except Exception as e:
        print(f"Error preparing model data: {e}")
        return
    
    # FIXED: Add debug prints to verify our data
    print(f"Target distribution: {np.unique(y, return_counts=True)}")
    
    # Train and evaluate model with REDUCED epochs for testing
    print("Training model...")
    model, history, X_market_test, X_sentiment_test, y_test = train_and_evaluate_model(
        X_market, X_sentiment, y, epochs=20, test_size=test_size
    )
    
    # FIXED: Get test predictions and examine them
    test_preds = model.predict([X_market_test, X_sentiment_test])
    print(f"Test predictions - shape: {test_preds.shape}")
    print(f"Test predictions - min: {np.min(test_preds)}, max: {np.max(test_preds)}")
    print(f"Test predictions - mean: {np.mean(test_preds)}")
    
    # If all predictions are very similar or all below threshold, adjust the threshold
    pred_mean = np.mean(test_preds)
    pred_std = np.std(test_preds)
    
    # FIXED: Dynamically adjust threshold based on prediction distribution
    if pred_std < 0.05:  # Very little variation in predictions
        print("WARNING: Predictions have very little variation!")
        suggested_threshold = max(0.5, pred_mean - pred_std)
        print(f"Suggested threshold based on prediction distribution: {suggested_threshold:.4f}")
    else:
        suggested_threshold = 0.6  # Default threshold
    
    # Implement trading strategy with vectorbt
    print("Backtesting trading strategy with vectorbt...")
    portfolio, backtest_data = implement_trading_strategy_vectorbt(
        model, 
        X_market_test, 
        X_sentiment_test, 
        combined_data,
        test_size=test_size,
        threshold=suggested_threshold  # Use the suggested or dynamically adjusted threshold
    )
    
    print("Backtest completed!")
    
    # FIXED: Try multiple thresholds with a wider range
    # Use more closely spaced thresholds around our suggested value
    low_threshold = max(0.4, suggested_threshold - 0.15)
    high_threshold = min(0.9, suggested_threshold + 0.15)
    thresholds = [low_threshold, suggested_threshold - 0.05, 
                 suggested_threshold, suggested_threshold + 0.05, high_threshold]
    
    results = {}
    
    for threshold in thresholds:
        print(f"\nTesting threshold: {threshold:.4f}")
        portfolio, _ = implement_trading_strategy_vectorbt(
            model, X_market_test, X_sentiment_test, combined_data, 
            test_size=test_size, threshold=threshold
        )
        
        # Create results dictionary
        results[threshold] = {
            'Total Return': portfolio.total_return(),
            'Sharpe Ratio': portfolio.sharpe_ratio(),
            'Max Drawdown': portfolio.max_drawdown()
        }
        
        # Add win_rate if available
        try:
            results[threshold]['Win Rate'] = portfolio.win_rate()
        except AttributeError:
            results[threshold]['Win Rate'] = "N/A"  # Not available in this version
    
    # Display comparative results
    results_df = pd.DataFrame(results).T
    print("\nComparison of Different Thresholds:")
    print(results_df)
    
    # Plot comparison of returns across thresholds
    plt.figure(figsize=(10, 6))
    plt.bar(results_df.index.astype(str), results_df['Total Return'])
    plt.title('Total Return by Confidence Threshold')
    plt.xlabel('Threshold')
    plt.ylabel('Total Return')
    plt.grid(axis='y', alpha=0.3)
    plt.show()

if __name__ == "__main__":
    main()
```
